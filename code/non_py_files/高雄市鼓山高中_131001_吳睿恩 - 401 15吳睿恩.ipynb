```python
import pandas as pd

# Path to your CSV file
file_path = '/content/week7 - 工作表1.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df
```

```python
Day = df[df['Meal'] == 'Breakfast']
```

```python
Day
```

```python
food_items = pd.concat([Day['Grains'], Day['Protein'], Day['Dairy'], Day['Vegetables']])
```

```python
# Count the frequency of each food item
food_count = food_items.value_counts().reset_index()

# Rename columns for clarity
food_count.columns = ['Food Item', 'Count']

# Sort the food items by count in descending order
food_count = food_count.sort_values(by='Count', ascending=False)
```

```python
import plotly as plt
import plotly.express as px
import plotly.graph_objects as go
# Create a bar chart based on the sorted food item frequency
fig = px.bar(food_count, x='Food Item', y='Count', title="Food Item Frequency (Sorted by Count)", labels={
    "Food Item": "Food Item",
    "Count": "Frequency"
})

# Display the plot
fig.show()
```
