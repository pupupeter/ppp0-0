```python
pip install networkx
import networkx as nx

def create_graph():
    G = nx.Graph()

    # 添加節點和邊，邊的權重可以代表距離或時間
    G.add_edge("A", "B", weight=5)
    G.add_edge("A", "C", weight=10)
    G.add_edge("B", "C", weight=3)
    G.add_edge("B", "D", weight=2)
    G.add_edge("C", "D", weight=1)
    G.add_edge("D", "E", weight=6)

    return G

def find_best_route(graph, start, end):
    # 使用 Dijkstra 演算法找到最佳路線
    length, path = nx.single_source_dijkstra(graph, start, end)
    return length, path

def main():
    # 創建圖
    travel_graph = create_graph()

    # 設定出發地和目的地
    start_location = "A"
    end_location = "E"

    # 找到最佳路線
    distance, route = find_best_route(travel_graph, start_location, end_location)

    # 輸出結果
    print(f"從 {start_location} 到 {end_location} 的最佳路線是: {route}，距離為: {distance}")

if __name__ == "__main__":
    main()
```
